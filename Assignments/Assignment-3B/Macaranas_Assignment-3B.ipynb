````python
import pandas as pd

students = pd.read_csv("studentprofiles.csv")
recweek = pd.read_csv("recweekdata.csv")

# data cleaning 

# course cleaning
course_replacements = [("^Legal Management|^BS LM|^LM","BS Legal Management"),
                       ("^Communications Technology|^BS CTM|^CTM","BS Communications Technology Management"),
                      ("^Information Technology Entrepreneurship|^BS ITE|^ITE","BS IT Entrepreneurship"),
                      ("^Management Engineering|^BS ME|^ME","BS Management Engineering"),
                      ("^Management|^BS MGT|^MGT","BS Management")]

for course,replacement in course_replacements:
    recweek.loc[:,"course"] = recweek.loc[:,"course"].str.replace(course, replacement, regex=True)


# birthmonth data cleaning 
birthmonth_replacements = [("January|Jan|1","January"),("February|Feb|2","February"),("March|Mar|3","March"),
                      ("April|Apr|4","April"),('May|5','May'),("June|Jun|6","June"),("July|Jul|7","July"),
                      ("August|Aug|8","August"),("September|Sep|9","September"),("January0|October|Oct|10","October"),
                      ("JanuaryJanuary|November|Nov|11","November"),("JanuaryFebruary|December|Dec|12","December")
                     ]

for month,replacement2 in birthmonth_replacements:
    recweek.loc[:,"birthmonth"] = recweek.loc[:,"birthmonth"].str.replace(month, replacement2, regex=True)

    
# birthyear data cleaning
recweek["birthyear"] = recweek["birthyear"].astype(str)
year_replacements = [(r"^97","1997"),(r"^98","1998"),(r"^99","1999")]

for year,replacement3 in year_replacements:
    recweek.loc[:,"birthyear"] = recweek.loc[:,"birthyear"].str.replace(year, replacement3, regex=True)


# phonenumber data cleaning 
phonenumber_replacements = [(r"^\+63 ","0"),(r"^9","09")]

for number,replacement4 in phonenumber_replacements:
    recweek.loc[:,"phonenumber"] = recweek.loc[:,"phonenumber"].replace(number, replacement4, regex=True)


# other lists
firstname =  students.loc[:,"firstname"]
lastname =  students.loc[:,"lastname"]
birthmonth =  recweek.loc[:,"birthmonth"]
birthday =  recweek.loc[:,"birthday"]
birthyear =  recweek.loc[:,"birthyear"]
course =  recweek.loc[:,"course"]
phonenumber =  recweek.loc[:,"phonenumber"]

# merge gender
recweek.drop("firstname",inplace=True,axis=1)
recweek = pd.merge (students, recweek, left_on=["lastname"], right_on=["lastname"])


#adding carrier
carrier = {"0917":"Globe","0918":"Smart", "0927":"Globe"}
recweek.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])
recweek["carrier"] = recweek.loc[:,"phonenumber"].str[0:4].apply(lambda x: carrier[x])


# creating a new csv
clean_df = pd.DataFrame(recweek)

clean_df.to_csv(r"cleanrecweekdata3b.csv", index=False)

````

````python
# number of people born each month
month_html = """"""
monthcount_html = """"""
months_order = ["January","February","March","April","May","June","July",
                "August","September","October","November","December"]
for month in months_order:
    birthmonth_count = clean_df.groupby(["birthmonth"])["firstname"].count()
    birthmonth = """
            <p> {month} </p>
            """ 
    monthcount =  """
            <p> {count} </p>
            """ 
    month_html += birthmonth.format(month=month) 
    monthcount_html += monthcount.format(count=birthmonth_count[month])

# number of people born each day
day_html = """"""
daycount_html = """"""
birthday_count = clean_df.groupby(["birthday"])["firstname"].count()
for day,count in birthday_count.iteritems():
    birthday = """
            <p> {day} </p>""" 
    daycount = """
            <p> {count} </p>""" 
    day_html += birthday.format(day=day)
    daycount_html += daycount.format(count=count)

# number of people born each year
birthyear_html = """"""
yearcount_html = """"""
birthyear_count = clean_df.groupby(["birthyear"])["firstname"].count()
for year,count in birthyear_count.iteritems():
    birthyear = """
            <p> {year} </p>"""
    yearcount = """
            <p> {count} </p>""" 
    birthyear_html += birthyear.format(year=year)
    yearcount_html += yearcount.format(count=count)

# number of people in each course
course_html = """"""
coursecount_html = """"""
course_count = clean_df.groupby(["course"])["firstname"].count()
for title,count in course_count.iteritems():
    course = """
            <p> {title} </p>""" 
    coursecount = """
            <p> {count} </p>""" 
    course_html += course.format(title=title)
    coursecount_html += coursecount.format(count=count)

# number of people per carrier
carrier_html = """"""
carriercount_html = """"""
carrier_count = clean_df.groupby(["carrier"])["firstname"].count()
for name,count in carrier_count.iteritems():
    carrier = """
            <p> {name} </p>""" 
    carriercount = """
            <p> {count} </p>"""
    carrier_html += carrier.format(name=name)
    carriercount_html += carriercount.format(count=count)
    
# number of people per carrier & gender
gcarrier_html = """"""
gcarriercount_html = """"""
gendercarrier_count = clean_df.groupby(["carrier","gender"])["firstname"].count()
for gcarrier, count in gendercarrier_count.iteritems():
    column1 = """
            <p> {} {} </p> """
    gcarriercount = """
            <p> {count} </p>"""
    gcarrier_html += column1.format(gcarrier[0],gcarrier[1])
    gcarriercount_html += gcarriercount.format(count=count)
    
````

````python
main_stat_html = structure_html.format(zero=style_html,one=month_html,two=monthcount_html,three=day_html,
                                       four=daycount_html,five=birthyear_html,six=yearcount_html,
                                       seven=course_html,eight=coursecount_html,nine=carrier_html,
                                       ten=carriercount_html, eleven=gcarrier_html, twelve=gcarriercount_html)

with open('stats3b.html',"w+") as file:
    file.write(main_stat_html)
````
