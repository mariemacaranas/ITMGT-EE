```python
import csv

recweek_list = list()

with open('recweekdata.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for line in csv_file:
        for row in csv_reader:
            if line_count == 0:
                pass
            else:
                recweek_list.append({"firstname":row[0], "lastname":row[1], "birthmonth":row[2], "birthday":row[3], "birthyear":row[4], "course":row[5], "phonenumber":row[6]})
            line_count += 1

recweek_list
```

```python
# data cleaning
import re


# course cleaning
course_list = [c["course"] for c in recweek_list]
course_replacements = [("^Legal Management|BS LM|LM","BS Legal Management"),
                       ("Communications Technology|BS CTM|CTM","BS Communications Technology Management"),
                      ("Information Technology Entrepreneurship|BS ITE|ITE","BS IT Entrepreneurship"),
                      ("^Management Engineering|BS ME|ME","BS Management Engineering"),
                      ("^Management|BS MGT|MGT","BS Management")]

for one,two in course_replacements:
    course_list = [re.sub(one, two, course) for course in course_list]
    

# birthmonth data cleaning --working
birthmonth_list = [c["birthmonth"] for c in recweek_list]
month_replacements = [("January|Jan|1","January"),("February|Feb|2","February"),("March|Mar|3","March"),
                      ("April|Apr|4","April"),('May|5','May'),("June|Jun|6","June"),("July|Jul|7","July"),
                      ("August|Aug|8","August"),("September|Sep|9","September"),("January0|October|Oct|10","October"),
                      ("JanuaryJanuary|November|Nov|11","November"),("JanuaryFebruary|December|Dec|12","December")
                     ]

for dirty,clean in month_replacements:
    birthmonth_list = [re.sub(dirty, clean, month) for month in birthmonth_list]


# birthmonth data cleaning --> some numbers are repeating ("1199977")????
birthyear_list = [c["birthyear"] for c in recweek_list]
years_replacement = [("^97","1997"),("^98","1998"),("^99","1999")]

for three,four in years_replacement:
    birthyear_list = [re.sub(three, four, year) for year in birthyear_list]
    

# phonenumber data cleaning
phonenumber_list = [c["phonenumber"] for c in recweek_list]
phone_replacement = [("^\+63 ","0"),("^9","09")]

for five,six in phone_replacement:
    phonenumber_list = [re.sub(five,six,number) for number in phonenumber_list]

# other lists
firstname = [c["firstname"] for c in recweek_list]
lastname = [c["lastname"] for c in recweek_list]
birthday = [c["birthday"] for c in recweek_list]

# new csv file 
clean_recweek_list = list()
with open('cleanrecweekdata.csv','w+') as clean_file:
    csv_writer = csv.writer(clean_file) # creating a variable in this line allows you to create rows
    
    csv_writer.writerow(['firstname','lastname','birthmonth','birthday','birthyear','course','phonenumber'])
    
    for i in range(len(firstname)):
        csv_writer.writerow([firstname[i],lastname[i],birthmonth_list[i],birthday[i],birthyear_list[i],course_list[i],phonenumber_list[i]])
    

```
